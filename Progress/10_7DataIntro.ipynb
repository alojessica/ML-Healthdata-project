```python
# trying to see if columnd with many nulls are rotating questions (not core)

optional_vars = ['ADSLEEP', 'ADPLEAS1', 'ADDOWN1', 'FEELNERV', 'STOPWORY',
                 'ECIGARET', 'ECIGNOW', 'MARIJAN1', 'USEMRJN2', 'RSNMRJN1']

# see how many missing per state
missing_by_state = df.groupby('_STATE')[optional_vars].apply(lambda x: x.isna().mean())

# show states where >95% are missing for each variable
high_missing = (missing_by_state > 0.95)
print(high_missing.sum(axis=0))  # how many states skipped each var
print(high_missing[high_missing.any(axis=1)])  # which states skipped which vars
```

```python
# new brfss dataset 2 (updated by removing redudancy and optional modules)

keep_cols = [
    # Outcomes / targets
    'MENTHLTH', '_MENT14D',

    # Demographics
    'SEX1', 'MARITAL', 'EDUCA', 'EMPLOY1', 'INCOME2', 'RENTHOM1', 'DECIDE',
    '_IMPRACE', '_AGEG5YR', '_EDUCAG', '_INCOMG', '_BMI5CAT',

    # Health access
    'HLTHPLN1', 'PERSDOC2', 'MEDCOST', 'CHECKUP1',

    # Lifestyle / risk
    'EXERANY2', 'SLEPTIM1',
    '_RFHLTH', '_PHYS14D', '_TOTINDA',
    '_SMOKER3', '_RFSMOK3',
    'DRNKANY5', '_RFBING5', '_RFDRHV6', '_DRNKWEK',

    # Geography to potentially use with SMART 
    '_STATE', '_METSTAT', '_URBSTAT', 'MSCODE'
]

# --- 2) columns to DROP

drop_cols = [
    # Raw redundancies replaced by calculated versions
    'GENHLTH', 'POORHLTH',
    'SMOKE100', 'SMOKDAY2',
    'ALCDAY5', 'AVEDRNK2', 'DRNK3GE5', 'DRNKDRI2',

    # Optional modules (very sparse across states in 2018)
    'ECIGARET', 'ECIGNOW',
    'MARIJAN1', 'USEMRJN2', 'RSNMRJN1',
    'ADSLEEP', 'ADPLEAS1', 'ADDOWN1', 'FEELNERV', 'STOPWORY',

    # Conditional lifetime smoking details (sparse)
    'COPDSMOK', 'LCSNUMCG',

    # Age redundancies (we’ll use _AGEG5YR)
    '_AGE_G', '_AGE65YR'
]


# (a) Start from the intersection so we don't KeyError if some cols are absent
existing_keep = [c for c in keep_cols if c in df.columns]
existing_drop = [c for c in drop_cols if c in df.columns]

# (b) Start with only the keep columns (preferred for modeling)
df2 = df[existing_keep].copy()

# (c) If any of the drop columns slipped in (e.g., because you also kept raw versions),
#     drop them here defensively (won’t error if none present).
df3 = df2.drop(columns=existing_drop, errors='ignore')




print("df3 shape:", df3.shape)
print(df3.head(3))

missing_counts = df3.isnull().sum()
print("\nMissing values per column:\n", missing_counts)
```
